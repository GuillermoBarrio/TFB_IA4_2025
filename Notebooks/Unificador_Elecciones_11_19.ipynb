Ahora almacenamos todos estos cambios en una sola función, que aplicaremos a los 5 datasets, incluyendo un formato inicial a la columna del nombre de municipio.

```python
def transformacion(df_eleccion, fecha):

  df_eleccion = df_eleccion[['Sección', 'cod_ccaa', 'cod_prov', 'cod_mun', 'cod_sec', 'CCAA',
       'Provincia', 'Municipio', 'Censo_Esc', 'Votos_Total', 'Participación',
       'Nulos', 'Votos_Válidos', 'Blanco', 'V_Cand', 'PP', 'PSOE', 'Cs', 'UP',
       'IU', 'VOX', 'UPyD', 'MP', 'CiU', 'ERC', 'JxC', 'CUP', 'DiL', 'PNV',
       'Bildu', 'Amaiur', 'CC', 'FA', 'TE', 'BNG', 'PRC', 'GBai', 'Compromis',
       'PACMA', 'Otros', '% PP', '% PSOE', '% UP', '% Vox', '% Cs', '% IU',
       'Ganador', 'Segundo', 'Tercero', 'Cuarto', 'Quinto', 'Población Total', 'Hombres', 'Mujeres',
       '% mayores 65 años', '% 20-64 años', '% menores 19 años',
       'Afiliados SS Minicipio', '% Afiliados SS autónomos',
       '% Afiliados SS / Población', 'Paro Registrado Municipio',
       '% Paro Hombres', '% Paro mayores 45',
       '% Paro s/ Afiliados SS Municipio', 'Renta persona 2017',
       'Renta persona 2015', 'Renta hogar 2017', 'Renta hogar 2015',
       'Renta Salarios 2018', 'Renta Salarios 2015', 'Renta Pensiones 2018',
       'Renta Pensiones 2015', 'Renta Desempleo 2018', 'Renta Desempleo 2015']]

  cols_integer = ['Censo_Esc', 'Votos_Total', 'Nulos', 'Votos_Válidos', 'Blanco', 'V_Cand', 'PP', 'PSOE', 'Cs', 'UP',
       'IU', 'VOX', 'UPyD', 'MP', 'CiU', 'ERC', 'JxC', 'CUP', 'DiL', 'PNV',
       'Bildu', 'Amaiur', 'CC', 'FA', 'TE', 'BNG', 'PRC', 'GBai', 'Compromis',
       'PACMA', 'Otros', 'Población Total', 'Hombres', 'Mujeres']

  cols_float = ['Participación', '% PP', '% PSOE', '% UP', '% Vox', '% Cs', '% IU',  '% mayores 65 años', '% 20-64 años', '% menores 19 años',
       'Afiliados SS Minicipio', '% Afiliados SS autónomos',
       '% Afiliados SS / Población', 'Paro Registrado Municipio',
       '% Paro Hombres', '% Paro mayores 45',
       '% Paro s/ Afiliados SS Municipio', 'Renta persona 2017',
       'Renta persona 2015', 'Renta hogar 2017', 'Renta hogar 2015',
       'Renta Salarios 2018', 'Renta Salarios 2015', 'Renta Pensiones 2018',
       'Renta Pensiones 2015', 'Renta Desempleo 2018', 'Renta Desempleo 2015']

  # List of columns to fill NaN values, defined based on original notebook's cols_na
  cols_to_fill_na = ['Censo_Esc', 'Votos_Total', 'Participación', 'Nulos', 'Votos_Válidos', 'Blanco',
       'V_Cand', 'PP', 'PSOE', 'Cs', 'UP', 'IU', 'VOX', 'UPyD', 'MP', 'CiU',
       'ERC', 'JxC', 'CUP', 'DiL', 'PNV', 'Bildu', 'Amaiur', 'CC', 'FA', 'TE',
       'BNG', 'PRC', 'GBai', 'Compromis', 'PACMA', 'Otros', '% PP', '% PSOE',
       '% UP', '% Vox', '% Cs', '% IU', 'Población Total', 'Hombres', 'Mujeres',
       '% mayores 65 años', '% 20-64 años', '% menores 19 años',
       'Afiliados SS Minicipio', '% Afiliados SS autónomos',
       '% Afiliados SS / Población', 'Paro Registrado Municipio',
       '% Paro Hombres', '% Paro mayores 45',
       '% Paro s/ Afiliados SS Municipio', 'Renta persona 2017',
       'Renta persona 2015', 'Renta hogar 2017', 'Renta hogar 2015',
       'Renta Salarios 2018', 'Renta Salarios 2015', 'Renta Pensiones 2018',
       'Renta Pensiones 2015', 'Renta Desempleo 2018', 'Renta Desempleo 2015']

  # Fill NaN values for all relevant numeric columns first
  for col in cols_to_fill_na:
    if col in df_eleccion.columns:
      df_eleccion[col] = df_eleccion[col].fillna(df_eleccion[col].mean(numeric_only=True))

  # Now, safely convert integer columns
  for col in cols_integer:
    df_eleccion[col] = df_eleccion[col].astype('int32')

  # And convert float columns
  for col in cols_float:
    df_eleccion[col] = df_eleccion[col].astype('float32')

  df_eleccion['Elecciones'] = fecha

  ccaa = { 'Andalucía' : 'Andalucía',
        'Aragón' : 'Aragón',
         'Melilla' : 'Melilla',
         'Ceuta' : 'Ceuta',
        'Cantabria' : 'Cantabria',
        'Asturias' : 'Asturias',
        "'Islas Baleares'": 'Islas Baleares',
         'Cataluña' : 'Cataluña',
        'Galicia' : 'Madrid',
        'Extremadura' : 'Galicia',
        'Castilla y León' : 'Castilla - La Mancha',
        'Castilla - La Mancha' : 'Castilla y León',
        'C. Valenciana' : 'Extremadura',
        'La Rioja' : 'C. Valenciana',
        'País Vasco' : 'La Rioja',
        'Murcia' : 'País Vasco',
        'Navarra' : 'Murcia',
        'Madrid' : 'Navarra',
        'Canarias' : 'Islas Canarias'
}
  df_eleccion['CCAA'] = df_eleccion['CCAA'].map(ccaa)

  df_eleccion['Distrito'] = df_eleccion['Sección'].str[16:18]

  # Reorder columns to include newly created ones
  final_cols_order = ['Elecciones', 'Sección', 'cod_ccaa', 'cod_prov', 'cod_mun', 'cod_sec', 'CCAA',
       'Provincia', 'Municipio', 'Distrito', 'Censo_Esc', 'Votos_Total', 'Participación',
       'Nulos', 'Votos_Válidos', 'Blanco', 'V_Cand', 'PP', 'PSOE', 'Cs', 'UP',
       'IU', 'VOX', 'UPyD', 'MP', 'CiU', 'ERC', 'JxC', 'CUP', 'DiL', 'PNV',
       'Bildu', 'Amaiur', 'CC', 'FA', 'TE', 'BNG', 'PRC', 'GBai', 'Compromis',
       'PACMA', 'Otros', '% PP', '% PSOE', '% UP', '% Vox', '% Cs', '% IU',
       'Ganador', 'Segundo', 'Tercero', 'Cuarto', 'Quinto', 'Población Total', 'Hombres', 'Mujeres',
       '% mayores 65 años', '% 20-64 años', '% menores 19 años',
       'Afiliados SS Minicipio', '% Afiliados SS autónomos',
       '% Afiliados SS / Población', 'Paro Registrado Municipio',
       '% Paro Hombres', '% Paro mayores 45',
       '% Paro s/ Afiliados SS Municipio', 'Renta persona 2017',
       'Renta persona 2015', 'Renta hogar 2017', 'Renta hogar 2015',
       'Renta Salarios 2018', 'Renta Salarios 2015', 'Renta Pensiones 2018',
       'Renta Pensiones 2015', 'Renta Desempleo 2018', 'Renta Desempleo 2015']

  df_eleccion = df_eleccion[final_cols_order]

  df_eleccion = df_eleccion.rename(columns = {'% Vox' : '% VOX'})

  df_eleccion['Municipio'] = df_eleccion['Municipio'].str.title()

  return df_eleccion
```
